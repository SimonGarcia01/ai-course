### Cual es la media de  `Customers_Age`?

```python
dfCustomerAge = sales["Customer_Age"]
```

Usar `.mean()`

```python
dfCustomerAge.mean()
```

Mostrar una grafica tipo: <b>density (KDE)</b> y un  <b>box plot</b> para`Customer_Age` :

```python
dfCustomerAge.plot(kind="kde", title="Density", figsize=(14,6))
```

```python
dfCustomerAge.plot(kind='box', title="Boxplot", vert=False, figsize=(16,4))
```
